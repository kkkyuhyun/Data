## 데이터탐색

```python
# income 범주화 cut() 사용 상 - 중 - 하로 구분
# labels = ["하", "중", "상"])
# level로 5번째 컬럼 만들기
min_value= df['income'].min()
max_value=df['income'].max()
print(f"income 데이터 최소값:{min_value}, 최대값: {max_value}")
bins=[min_value,(min_value+max_value)/3, (min_value+max_value)*2/3, max_value]
labels=['하','중','상']
df['level']=pd.cut(df['income'],bins=bins,labels=labels,include_lowest=True)
print(df[['income','level']])
df
```

```python
# 범주형 중복없이 value 추출
t2 = df.level.unique()
t2
```

```python
# 원하는 데이터만 
# level 에서 '상'데이터만 출력
t2 = df[df.level.isin(['상'])] 
t2
```

```python
# '상'데이터 제외 필터링
t3 = df[~df.level.isin(['상'])] # 원하는 데이터 제외 필터링
t3
```

```python
# groupby ['default', 'student'] income의 평균값 구하기

average_income=df.groupby(['default','student'])['income'].mean()
print(average_income)
df
```

```python
# yes는 1로, no는 0으로
# replace({'Yes':1, 'No':0})
df['default']=df['default'].replace({'Yes':1,'No':0})
#교수님 코드
# df=df.replace({'Yes':1, 'No':0})
#print(df)
#df['default']=df['default'].map({0:'no',1:'yes'})
#print(df)

# 피봇 테이블
# 'default' 평균 / index=['student', 'level']

pivot_table = df.pivot_table(['default'], index=['student','level'],aggfunc='mean')
print(pivot_table)
```
