## 이항분포

```python
from scipy import stats
# stats.binom.pmf(x,n,p) - 질량함수
# stats.binom.cdf(x,n.p) - 누적함수
```

### 주사위를 4회 던져서 6의 눈이 2회 나올 확률은?

```python
stats.binom.pmf(2,4,1/6)
```

### 예제1) 불량품이 0.1인 공정에서 15개의 제품을 추출하였을 때, 다음의 확률을 구하여라.
### P(X=2), P(X<=2), P(X>=10)



```python
#★시험출제 예정
#p=0.1
#n=15
#x=2(경우의 수)
print(stats.binom.pmf(2,15,1/10)) #p(x=2)
print(stats.binom.cdf(2,15,1/10)) #p(x<=2)
print(1- stats.binom.cdf(9,15,1/10)) #p(>=10)
```

### 예제2) 동전을 다섯 번 던졌을 때 앞면이 세번 나올 확률과 세번 이하로 나올 확률 


```python
#n=5, p=1/2, x=3 (x,n,p)
print(stats.binom.pmf(3,5,1/2))
print(stats.binom.cdf(3,5,1/2)) 
```

## 이항분포 그래프그리기

```python
import numpy as np
import matplotlib.pyplot as plt
n=20
x=np.arange(21)
ps=np.arange(1,10)/10

for p in ps:
    plt.plot(x, stats.binom.pmf(x, n, p), label='(p=' + str(p) + ')')
    
plt.xlabel('X')
plt.ylabel('P(X)')
plt.grid()
plt.legend()
plt.show()
```

## p = 0.5 일때 누적 그래프 그리기

```python
import matplotlib.pyplot as plt
from scipy import stats

x = np.arange(21)

fig = plt.figure()

ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)


ax1.plot(x, stats.binom.pmf(x,20,0.5))
ax2.plot(x, stats.binom.cdf(x,20,0.5))
     
plt.show()
```

## 포아송분포 그래프그리기

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

x = np.arange(11)

fig = plt.figure()

ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

for i in range(1, 10):
    ax1.plot(stats.poisson.pmf(x,i))

ax2.plot(stats.poisson.cdf(x,3))
     
plt.show()
```

## 정규분포곡선 그리기

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

x = np.arange(100)

fig = plt.figure()

ax1 = fig.add_subplot(2, 3, 1)
ax2 = fig.add_subplot(2, 3, 2)
ax3 = fig.add_subplot(2, 3, 3)
ax4 = fig.add_subplot(2, 3, 4)
ax5 = fig.add_subplot(2, 3, 5)
ax6 = fig.add_subplot(2, 3, 6)

ax1.plot(stats.norm.pdf(x-50,0,10))
ax2.plot(stats.norm.pdf(x-50,0,20))
ax3.plot(stats.norm.pdf(x-50,0,30))
ax4.plot(stats.norm.pdf(x-60,0,10))
ax5.plot(stats.norm.pdf(x-50,0,10))
ax6.plot(stats.norm.pdf(x-40,0,10))
   
plt.tight_layout()

plt.show()
```

## 평균: 390, 표준편차: 25
### P(X>=390)
### P(354<=X<=430)
### P(X>=430)
### P(X<=360)+P(X>=430)
### 상위 3% 이내에 들 점수


```python
#Q1. 평균 390 이상일 확률
print(1-stats.norm.cdf(390,390,25))
#Q2. 354이상 430이하일 확률
print(stats.norm.cdf(430,390,25) - stats.norm.cdf(354,390,25))
#Q3. 430이상일 확률
print(1-stats.norm.cdf(430,390,25))
#Q4. 360이하이거나 430이상일 확률
print(1-stats.norm.cdf(430,390,25)+stats.norm.cdf(360,390,25))
#Q5. 상위 3%이내에 들 점수는?
print(stats.norm.ppf(0.97,loc=390,scale=25)) #scale 표준편차 
```

## 정규분포의 랜덤 샘플숫자

```python
from scipy import stats
import matplotlib.pyplot as plt

rv=stats.norm(0,5)
z=rv.rvs(size=10000)
plt.hist(z,bins=100)
plt.show()
```

각각 1000명
남자/여자 키, 몸무게, 100m 달리기 – (평균/표준편차)
각각 히스토그램 만들기
남/녀 키, 몸무게, 100m달리기 –평균, 중간값, 표준편차, 최댓값, 최솟값 구하기


```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np

man_height_rv = stats.norm(loc = 178, scale = 20)
woman_height_rv = stats.norm(loc = 165, scale = 20)

man = man_height_rv.rvs(size=1000)
woman = woman_height_rv.rvs(size=1000)

fig = plt.figure()

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)


ax1.hist(man, bins=10 )
ax2.hist(woman, bins = 10)

plt.show()

#평균값, 중앙값, 최대값, 최소값, 최빈값 뽑아내기
#pandas -> describe() 
#numpy -> 숫자를 다루는 라이브러리 array 기술통계 sin cos 
# np.mean() ...
print(np.max(man)), np.median((man)), np.mean((man)), np.std((man))
```
